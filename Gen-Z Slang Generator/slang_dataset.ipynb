```python
import pandas as pd
```

```python
emoji_slang = pd.read_csv("/content/drive/MyDrive/MLB_Kaggle_Gemma/genz_emojis.csv")
word_slang = pd.read_csv("/content/drive/MyDrive/MLB_Kaggle_Gemma/genz_slang.csv")
```

```python
emoji_slang.head()
```

```python
emoji = pd.read_csv("/content/drive/MyDrive/MLB_Kaggle_Gemma/genz_emojis.csv")
emoji.rename(columns={'emoji':'Slang'}, inplace=True)
emoji['Description'] = emoji['Name'] +', ' + emoji['Description']
emoji.drop(columns='Name', inplace=True)
emoji_slang = emoji
emoji_slang.head()
```

```python
word_slang.head()
```

```python
word_slang.rename(columns={'keyword':'Slang', 'description':'Description'}, inplace=True)
word_slang.head()
```

```python
word_slang2 = pd.read_csv("/content/drive/MyDrive/MLB_Kaggle_Gemma/slang.txt", delimiter = '=')
word_slang2.to_csv("slang.csv", index = False)
word_slang2.rename(columns={'AFAIK':'Slang', 'As Far As I Know':'Description'}, inplace=True)
word_slang2.head()
```

```python
all_slang = pd.concat([emoji_slang, word_slang, word_slang2], ignore_index=True)
all_slang.to_csv("all_slang.csv", index=False)
all_slang
```

```python
abbervation_slang = pd.read_csv("/content/drive/MyDrive/MLB_Kaggle_Gemma/abbrevations.csv")
abbervation_slang.rename(columns={'?':'Slang', 'I have a question':'Description'}, inplace=True)
abbervation_slang.head()
```

```python
all_slang = pd.concat([all_slang, abbervation_slang], ignore_index=True)
all_slang
```

```python
all_slang.to_csv("all_slang.csv", index=False)
```
